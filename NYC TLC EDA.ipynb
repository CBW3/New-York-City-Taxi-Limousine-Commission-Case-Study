```python
#importing relevant libraries & packages for analysis 
import pandas as pd
import matplotlib as plt
import numpy as np
import datetime as dt
import seaborn as sns 
```

```python
#loading in & viewing the dataset 
df=pd.read_csv('2017_Yellow_Taxi_Trip_Data.csv')
df.head(10)
```

```python
#getting a feel for the size of the dataset
df.size
```

```python
#further exploring the dataset through summary statistics table 
df.describe()
```

```python
#exploring data types in the dataset
df.info()
```

```python
#converting relevant columns to datetime format
df['tpep_pickup_datetime']=pd.to_datetime(df['tpep_pickup_datetime'])
df['tpep_dropoff_datetime']=pd.to_datetime(df['tpep_dropoff_datetime'])
```

```python
#crafting box plot to visually inspect for potential outliers for trip distance column
sns.boxplot(data=None, x=df['trip_distance'], fliersize=1)
```

```python
#crafting box plot to visually inspect for potential outliers for total amount column 
sns.boxplot(x=df['total_amount'], fliersize=1)
```

```python
#removing trips with costs associated with trip distance of 0, doesn't make any sense
df_2 = df['trip_distance'].loc[~(df==0).all(axis=1)]
```

```python
#constructing scatterplot of total amount & trip distance relationship 
sns.scatterplot(x=df['total_amount'], y=df_2)
```

```python
#grouping trip distance by DOLocationID
df.groupby('DOLocationID')['trip_distance'].mean()
```

```python
#creating bar chart visualization to explore relationship between location and trip distance 
sns.barplot(data=df, x=df['DOLocationID'], y=df['trip_distance'])
```
